## 1. Statics: 2D Truss Analysis (Method of Joints)

We have a simple 3-member triangular truss loaded at the apex. We will calculate the reaction forces and internal member forces.

```python
import sympy as sp

# Define unknowns
Rx_A, Ry_A, Ry_B = sp.symbols('Rx_A Ry_A Ry_B')

# Given external force at apex (Node C) in Newtons
F_C = 500 
L = 10 # Base length in meters

# Equilibrium Equations (Sum of forces = 0, Sum of moments about A = 0)
eq_fx = sp.Eq(Rx_A, 0)
eq_fy = sp.Eq(Ry_A + Ry_B - F_C, 0)
eq_ma = sp.Eq(Ry_B * L - F_C * (L / 2), 0)

# Solve for reactions
reactions = sp.solve((eq_fx, eq_fy, eq_ma), (Rx_A, Ry_A, Ry_B))
print(f"Reaction Forces: {reactions}")
```

## 2. Thermodynamics: Ideal Gas Law Expansion

Calculating the final temperature of a gas undergoing an isentropic expansion (reversible adiabatic) process.

```python
import numpy as np

# Initial State
T1 = 300 # K
P1 = 101.3 # kPa
P2 = 50.0 # kPa
gamma = 1.4 # specific heat ratio for air

# Isentropic relation: T2 = T1 * (P2/P1)^((gamma - 1)/gamma)
T2 = T1 * ((P2 / P1) ** ((gamma - 1) / gamma))

print(f"Final Temperature after isentropic expansion to 50 kPa: {T2:.2f} K")
```
